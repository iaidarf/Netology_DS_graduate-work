# Рекомендательная система

**Рекомендательная система** (РС) - это система, которая использует алгоритмы машинного обучения для анализа данных о предпочтениях и поведении пользователей, чтобы предлагать им наиболее подходящие товары, услуги или контент. РС используются в различных областях, включая электронную коммерцию, медиа, социальные сети, музыку и видео. 

Применение РС имеет множество практических преимуществ:

1. *Увеличение продаж*. Рекомендации товаров и услуг, которые наиболее подходят конкретному пользователю, могут увеличить вероятность покупки и, следовательно, увеличить продажи.

2. *Улучшение пользовательского опыта*. Рекомендации помогают пользователям быстрее и легче найти то, что им нужно, улучшает их общий опыт использования сайта или приложения.

3. *Увеличение лояльности клиентов*. Если пользователь получает рекомендации, которые действительно соответствуют его предпочтениям, он, вероятнее всего, будет возвращаться и делать покупки снова.

4. *Экономия времени*. Рекомендации позволяют пользователям экономить время, потому что они не тратят много времени на поиск товаров или услуг.

5. *Повышение конкурентоспособности*. Компании, использующие РС, могут повысить свою конкурентоспособность, предлагая более персонализированный и удобный сервис для своих клиентов.

# Типы РС

## 1. Popularity-based/Рекомендации на основе популярности

Особенности:
- Наиболее простая система: чем выше средний рейтинг фильма/товара/статьи, тем вероятнее, что система будет рекомендовать именно их.
- Недостаток: не учитываются предпочтения конкретного пользователя.

## 2. Content-based/Рекомендации на основе содеражания

Особенности:
- Простой и понятный:  рекомендуются товары/услуги похожие на те, которые пользователь приобретал ранее 
- Недостаток: покупатели не пробуют новое
- Нужно много признаков объекта
- Требуется feature engineering
- Нет проблемы холодного старта для объектов
- Есть проблема холодного старта для пользователя


## 3. Collaborative filtering/Коллаборативная фильтрация

Особенности:
- Используются только результаты взаимодействий
- Не нужно налаживать процесс извлечения фичей
- Проблема холодного старта (новый юзер/айтем - не ясно, что рекомендовать)
- Нужно много данных
- Предсказать значения потенциального взаимодействия

Два способа:

**User-based/похожесть пользователей**
- находят близких по предпочтениям пользователей и рекомендуют одному из них то, что уже попробовал другой
- лучше работает, когда объектов больше, чем пользователей
- недостаток: с увеличением количества пользователей линейно увеличивается сложность вычисления рекомендации

**Item-based/похожесть объектов**
- сравнивается близость товаров/услуг
- лучше работает, когда пользователей больше, чем объектов
- объекты добавляются реже, чем пользователи, поэтому этот способ менее затратен с точки зрения перерасчета модели


## 4. Рекомендации на основе скрытых факторов

Особенности:
- эмбеддинги пользователя и объекта
- снижение размерности (SVD, ALS) позволяет решить проблему разреженных данных
- низкая интепретируемость результатов работы модели

## Матрица взаимодействий/предпочтений item-user

В центре РС находится так называемая **матрица предпочтений**. Это матрица, по одной из осей которой находятся все пользователи (users), а по другой – объекты (items). На пересечении некоторых пар (user, item) - известные показатели заинтересованности пользователя.

Эти пользовательские оценки можно получить двумя способами:

- явно (**explicit ratings**) – пользователь сам ставит рейтинг, оставляет отзыв, ставит лайкает и пр.
- неявно (**implicit ratings**) – пользователь явно свое отношение не выражает, но можно сделать косвенный вывод из его действий: купил товар – значит он ему нравится и т.п.

В нашем случае по одной оси будут фильмы, по другой - пользователи, значение на пересечении - % просмотра. Эта оценка является неявной, можно предположить, что если пользователь посмотрел фильм полностью (значение близкое к 100%), то он ему понравился. Если значение ближе к нулю, то скорее всего, фильм ему не понравился. Остается, конечно, вероятность что его отвлекли от просмотра и он уже к нему не вернулся.

Каждый пользователь, естественно, просмотрел лишь небольшую часть фильмов. И **задача рекомендательной системы предсказать отношение клиента к другим фильмам, про которые пока ничего не известно**. Другими словами нужно заполнить все незаполненные ячейки в матрице item-user.


# Метрика качества РС

Метрикой качества работы РС выберем **Mean Average Precision**. Рассмотрим как получается эта метрика поэтапно.

1. **p@K** - Precision at K (точность модели)\
$$p@K= \frac{TP}{K}$$
TP - количество релевантных объектов (объекты, которые выдала РС и с которыми провзаимодействовал пользователь)\
К - общее количество рекомендаций

2. **ap@K** - Average precision at K (средняя точность). Эта метрика, в отличие от первой, учитывает на каком месте стоит рекомендованный объект (чем выше, тем лучше). ap@K равна сумме p@i (где i - индекс от 1 до K) только для релевантных элементов, деленной на количество релевантных элементов:
$$ap@K(user)= \frac{1}{c_{user}} \sum_{i=1}^K p@i * rel_i$$
$c_{user}$ - кол-во релевантных объектов у пользователя \
$rel_i$ - индикатор релевантности, который позволяет учитывать только релевантные объекты (если рекомендованный объект релевантен, то $rel_i = 1$, иначе $rel_i = 0$)

3. **map@K** - Mean average precision at K (усредненная средняя точность). В этой метрике происходит переход от одного пользователя ко всем. Идея заключается в том, чтобы посчитать ap@K для каждого пользователя и усреднить:
$$map@K= \frac{1}{N} \sum_{i=1}^N ap@k(user_i)$$
N - количество пользователей.
